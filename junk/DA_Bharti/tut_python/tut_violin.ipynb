```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# violin plot
np.random.seed(42)
data1 = np.random.normal(0, 1,100)
data2 = np.random.normal(2, 1.5,100)
data3 = np.random.normal(-2, 0.5,100)

all_data = [data1, data2, data3]

plt.violinplot(all_data, showmedians=True)

plt.show()
```

```python
# violin plot
np.random.seed(42)
data1 = np.random.normal(0, 1,10)
data2 = np.random.normal(2, 1.5,10)
data3 = np.random.normal(-2, 0.5,10)

print(f"data1: {data1}")
print(f"data2: {data2}")
print(f"data3: {data3}")

categories = ["apple", "banana", "kiwi"]
all_data =   [data1, data2, data3]

plt.violinplot(all_data, showmedians=True)
plt.xlabel('Categories')
plt.ylabel('Value')
plt.title("Bivariate violin plot")

# Set category labels
plt.xticks([1, 2, 3], categories)

plt.show()
```
