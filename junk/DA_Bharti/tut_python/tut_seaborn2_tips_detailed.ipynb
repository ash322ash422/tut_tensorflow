```python
sns.countplot(data=df, x = 'sex', hue='size', palette='viridis') # TODO do this for sex, smoker, day, time, size
plt.show()
```

```python
# catplot
sns.catplot(data=df, x='day', hue='sex', palette='viridis', kind='count') # TODO kind='count' chages it to countplot
plt.show()
```

```python
# facet
sns.catplot(data=df, x='sex', hue='smoker', col='day', kind='count') # TODO kind='count' chages it to countplot
plt.show()

#obs: On fri, more males were smoking, and also more females were smoking.
# on Thurs, Sun, there is huge difference between smoker and non-smoker
# On sat, the smoker and non-smoker are quite similar
```

```python
# facet
sns.catplot(data=df, x='sex', hue='size', col='day', kind='count') # TODO kind='count' chages it to countplot
plt.show()

# On Thur,Sat,Sun: no single male
# on Fri, Sun:     no single female 
# on Fri: not group of 6 people,etc
```
